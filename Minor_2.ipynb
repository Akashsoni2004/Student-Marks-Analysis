```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("student_marks.csv")

#------------------------------------------------------------------
# Calculate average, max, min marks per subject for each student
subjects_marks = df.groupby("Subject")["Grand Total (out of 200)"]
stats = subjects_marks.agg(["mean", "max", "min"])
print(stats)


# Total marks of each student across all subjects
student_totals = df.groupby("Student_ID")["Grand Total (out of 200)"].sum().reset_index()

# bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x="Student_ID", y="Grand Total (out of 200)", hue="Student_ID", data=student_totals, palette="viridis")
plt.title("Comparison of Students by Total Marks")
plt.ylabel("Total Marks (out of 1000)")  # 5 subjects  200 max each
plt.xlabel("Student ID")
plt.show()

# Subject-wise comparison of students
plt.figure(figsize=(14, 8))
sns.barplot(x="Subject", y="Grand Total (out of 200)", hue="Student_ID", data=df, palette="Set2")
plt.title("Subject-wise Comparison of Students")
plt.ylabel("Marks (out of 200)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
